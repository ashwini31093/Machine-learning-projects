```python
#4.a.Do charges of people who smoke differ significantly from the people who don't?
#plot smoker vs charges plot

sns.barplot(x='smoker',y='charges', data=insurance_csv)
```

```python
#Hypothesis testing
#Split the data set into smokers and non-smoker charges
smoker_charges =insurance_csv[['smoker', 'charges']] 
smoker_charges=smoker_charges.groupby(by=["smoker"])
smoker_charges_yes=smoker_charges.get_group("yes")

smoker_charges_no=smoker_charges.get_group("no")
smoker_charges_yes=smoker_charges_yes.drop('smoker',axis=1)
smoker_charges_no=smoker_charges_no.drop('smoker',axis=1)
print('Smoker charges:',smoker_charges_yes)
print('Non-smoker charges:',smoker_charges_no)
```

```python
#Null Hypothesis: charges(smoker)=charges(non smoker)
#Alternate Hypothesis: charges(smoker)!=charges(non smoker)
#significance level  = 0.05


from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro
from statsmodels.stats.power import ttest_power

t_statistic, p_value = ttest_ind(smoker_charges_yes, smoker_charges_no)
print(t_statistic, p_value)
```

```python
# Here  p_value < 0.05 
```

```python
#Inference: 
#From above barplot, we can observe that ,for smokers the charges is around 32000 and for non-smokers it is around 7000.
#Hence the charges vary significantly upto 5 times from non-smokers and smokers.
# Also we got p_value<0.05, hence we reject null hypothesis and  charges for smokers differ from charges of non smoker.
```

```python
#4.b.Does bmi of males differ significantly from that of females?
#plot sex v/s BMI plot
sns.stripplot(insurance_csv['sex'],insurance_csv['bmi'],jitter=True);
```

```python
#Inference from plot: From the above plot we can infer that BMI for female and male doesn't vary much.
```

```python

#Split into sex and bmi
sex_bmi =insurance_csv[['sex', 'bmi']] 
sex_bmi=sex_bmi.groupby(by=["sex"])
female_bmi=sex_bmi.get_group("female")

male_bmi=sex_bmi.get_group("male")
male_bmi=male_bmi.drop('sex',axis=1)
female_bmi=female_bmi.drop('sex',axis=1)
print('male bmi:',male_bmi)

print('Female BMI:',female_bmi)
```

```python
#Null Hypothesis: bmi(male)= bmi(female) 
#Alternate Hypothesis:bmi(male)!= bmi(female)
#significance level  = 0.05
t_statistic, p_value = ttest_ind(male_bmi, female_bmi)
print(t_statistic, p_value)
```

```python
#Inference:
#since p-value 0.08>0.05 we accept Null hypothesis that bmi of male is almost same as that of female.
```
